Check for missing values in the Iris dataset

Fill missing values with the mean

**2. Task 2**

Plot histogram for “petal length (cm)”

**3. Task 3**

Plot boxplots of “sepal width (cm)” for each species

```python
# Import pandas for data handling
import pandas as pd
# Import numpy for numerical operations
import numpy as np
# Import seaborn for advanced plotting
import seaborn as sns
# Import matplotlib for plotting graphs
import matplotlib.pyplot as plt
# Import a dataset (Iris) from scikit-learn
from sklearn.datasets import load_iris


# Load the iris dataset into a variable
iris = load_iris()
print(iris)

# Create a DataFrame from the iris data
df = pd.DataFrame(iris.data, columns=iris.feature_names)
print(df)

# Add the target (0,1,2) as a new column
df['target'] = iris.target

# Map numeric target values to species names
df['species'] = df['target'].map({0:'setosa', 1:'versicolor', 2:'virginica'})

# Introduce some missing values intentionally in sepal length column
df.loc[5:10, 'sepal length (cm)'] = np.nan

# Display count of missing values in each column
print(df.isnull().sum())

# Calculate the median of sepal length (excluding NaN values)
median_value = df['sepal length (cm)'].median()

# Fill missing values in sepal length column with median value
df['sepal length (cm)'] = df['sepal length (cm)'].fillna(median_value)

# Print missing values count again to confirm replacement
print("\nAfter filling missing values:")
print(df.isnull().sum())

# Plot a histogram of sepal length
plt.hist(df['sepal length (cm)'], bins=15, color='lightgreen')
plt.title("Histogram of Sepal Length")
plt.show()

# Plot a boxplot of sepal length grouped by species
sns.boxplot(x='species', y='sepal length (cm)', data=df)
plt.show()

# Plot a scatter plot of sepal length vs petal length colored by species
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=df)
plt.show()

# Compute correlation between numeric features
corr = df.iloc[:, :-2].corr()

# Plot the correlation heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

**1. Task 1**

Check for missing values in the Iris dataset

Fill missing values with the mean

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
print(iris)

df = pd.DataFrame(iris.data, columns=iris.feature_names)
print(df)

df['target'] = iris.target

df['species'] = df['target'].map({0:'setosa', 1:'versicolor', 2:'virginica'})

df.loc[5:10, 'sepal length (cm)'] = np.nan

print(df.isnull().sum())

median_value = df['sepal length (cm)'].median()

df['sepal length (cm)'] = df['sepal length (cm)'].fillna(median_value)

print("\nAfter filling missing values:")
print(df.isnull().sum())
```

```python
plt.hist(df['petal length (cm)'], bins=15, color='skyblue', edgecolor='gray')
plt.title("Histogram of Petal Length")
plt.show()
```

```python
sns.boxplot(x='species', y='sepal width (cm)', data=df)
plt.show()
```
